```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/kartikjindgar/NYU-DataScience-Bootcamp-Fall23/main/Week4/data/boston_housing_data.csv'
housing = pd.read_csv(url)
```

```python
plt.style.use('bmh')
plt.figure(figsize=(12,8))
plt.plot(housing['ZN'],color='green',label='ZN')
plt.plot(housing['INDUS'],color='blue',linestyle='--',label='INDUS')
plt.title('Line Plot of ZN and INDUS')
plt.legend()
plt.show()
```

```python
data1 = {'X': np.random.rand(100)}
data2 = { 'Y': np.random.rand(100) + 0.5 * data1['X'] }
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
plt.scatter(df1, df2)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
data1 = {'X': np.random.rand(100)}
data2 = { 'Y': np.random.rand(100) - 0.5 * data1['X'] }
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
plt.scatter(df1, df2)
plt.xlabel('X')
plt.ylabel('Y')
```
